```python
"""Descargamos las stopwords"""
_SYMBOLS= set(";:,.\\-\"'/()[]¿?¡!{}~<>|\t\n\r")
#obtiene la lista de stopwords del paquete NLTK para español}
import nltk
_STOPWORDS=nltk.download('stopwords')
stop_words=stopwords.words('spanish')
```

```python
def normalize_text(text,
                   punct=False,
                   accents=False,
                   max_dup=2):
    """
    Estandariza el texto, remueve símbolos duplicados, acentos y puntuación
    """
    nfkd_f = unicodedata.normalize('NFKD', text)
    n_str = []
    c_prev = ''
    cc_prev = 0
    for c in nfkd_f:
        if not punct:
            if c in _SYMBOLS:
                continue
        if not accents and unicodedata.combining(c):
            continue
        if c_prev == c:
            cc_prev += 1
            if cc_prev >= max_dup:
                continue
        else:
            cc_prev = 0
        n_str.append(c)
        c_prev = c
    return "".join(n_str)
```

```python
# Normalizar stopwords con el mismo tratamiento de los textos  
_NORM_STOPWORDS=[normalize_text(W)  for W in stop_words]
```

```python
import pandas as pd
import math
import unicodedata
from collections import Counter
from nltk.corpus import stopwords
spanish_tokenizer = nltk.data.load('tokenizers/punkt/PY3/spanish.pickle') 
from nltk.stem.porter import PorterStemmer
```

```python
_SYMBOLS= set(";:,.\\-\"'/()[]¿?¡!{}~<>|\t\n\r—«»_“")
#obtiene la lista de stopwords del paquete NLTK para español
import nltk
nltk.download('stopwords')
_STOPWORDS=set(stopwords.words('spanish'))
print(_STOPWORDS)
stemmer = PorterStemmer()
```

```python
def normalize_text(text,
                   punct=False,
                   accents=False,
                   max_dup=2):
    """
    Estandariza el texto, remueve símbolos duplicados, acentos y puntuación
    """
    nfkd_f = unicodedata.normalize('NFKD', text)
    n_str = []
    c_prev = ''
    cc_prev = 0
    for c in nfkd_f:
        if not punct:
            if c in _SYMBOLS:
                continue
        if not accents and unicodedata.combining(c):
            continue
        if c_prev == c:
            cc_prev += 1
            if cc_prev >= max_dup:
                continue
        else:
            cc_prev = 0
        n_str.append(c)
        c_prev = c
    return "".join(n_str)
```

```python
# Normalizar stopwords con el mismo tratamiento de los textos  
_NORM_STOPWORDS = [normalize_text(sw)  for sw in _STOPWORDS]
```

```python
def filter_words(text, remove_stopwords=True):
    """ 
    Filtra los tokens que inician con los símbolos indicados
    """
    wlist = text.split()
    o_text = []
    # filtrado de palabras
    for t in wlist:
        if remove_stopwords and t in _NORM_STOPWORDS:
            continue
        if t.isnumeric():
            continue
        #Aplica stemming a los tokens
        t = stemmer.stem(t)
        o_text.append(t)
    return " ".join(o_text)
```

```python
import re
```

```python
def preprocesa(text):
	text = text.lower()
	text = normalize_text(text)
	text = filter_words(text)
	text = re.sub('http\S+', ' ', text)
	text = re.sub('@\S+', '', text)
	text = re.sub('secretaria', '', text)
	text = re.sub('dependencia', '', text)
	text = re.sub('solicitud', '', text)
	text = re.sub('solicito', '', text)
	text = re.sub('transparencia', '', text)
	text = re.sub('informacion', '', text)
	text = re.sub('ley', '', text)
	text = re.sub('publica', '', text)
	text = re.sub('articulo', '', text)
	text = re.sub('acceso', '', text)
	return text
```
